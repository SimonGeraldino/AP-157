```python
# Imports the important libraries
import networkx as nx
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import community as community_louvain
```

#### PART 1 - Network generation and characteristics *(30 points)*

1. Using the Python package networkx, define a network that represents the social network of friends in the diagram. *(15 points)*  

2. Calculate the ff. network characteristics: *(15 points; 5 points each)*
- Diameter
- Density
- Average Degree

3. Create the degree distribution plot and briefly discuss what it tells us about this social network. *(10 points)*

```python
# Creates an empty graph
G = nx.Graph()

# Creates the list of friends/people
friends = ["Bob", "Mike", "John", "Leah", "Jill", "Shane", "Emma", "Liz", "Allen", "Lisa"]

# Adds the list of people in the array "friends" as nodes
G.add_nodes_from(friends)

# Creates the list of friendships/connections
friendships = [("Bob", "John"), ("Bob", "Jill"), ("Bob", "Emma"), ("Bob", "Mike"),
               ("Mike", "Emma"), ("Mike", "Jill"),
               ("John", "Jill"), ("John", "Leah"), ("John", "Shane"),
               ("Leah", "Jill"), ("Leah", "Shane"),
               ("Jill", "Emma"), ("Jill", "Shane"),
               ("Emma", "Liz"), ("Emma", "Shane"),
               ("Shane", "Liz"),
               ("Liz", "Allen"),
               ("Allen", "Lisa")]

# Adds all edges from the list of connections seen in the array "friendships"
G.add_edges_from(friendships)

# Creates the network
nx.draw(G,
        with_labels = True,
        node_color = "lightblue",
        edge_color = "lightblue",
        node_size = 300,
        font_size = 15)
```

We can observe here are the nodes (people) and their connections to one another, based on the given sample social network diagram.

```python
# Gets the diameter of the network
diameter = nx.diameter(G)

# Gets the density of the network
density = nx.density(G)

# Gets the average degree of the network
average_degree = 2*nx.number_of_edges(G)/nx.number_of_nodes(G)

print("The diameter of the network is " + str(diameter) + ".")
print("The density of the network is " + str(density) + ".")
print("The average degree of the network is " + str(average_degree) + ".")
```

The diameter of the network is the shortest number of connections/links needed to connect the farthest nodes in the network.

The density of the network describes the ratio of number of links that are created in the network, with respect to the maximum possible links that network can have, based on the number of nodes it has.

Lastly, the average degree of the network is computed by

\begin{equation}
  \text{average degree} = \frac{2*(\text{no. of edges})}{\text{no. of nodes}}
\end{equation}

```python
# Creates the degree distribution plot for the network
degrees = [degree for node, degree in G.degree()]

# Sets the parameters of the bar graph
plt.hist(degrees, bins = range(1, max(degrees) + 2), align = 'left', edgecolor = 'black')
plt.title("Degree Distribution for this network")
plt.xlabel("Degree")
plt.ylabel("Frequency")

# Shows the plot
plt.show()
```

The degree distribution plot shows the degree of connection people have and the frequency of it appearing inside the network. From observing the network, we can say that there are:

*   1 person who is only connected another person (Lisa)
*   1 person who is connected to the most people (Jill)



#### PART 2 - Centrality Measures *(30 points)*

Generate a table of centrality measures per node. Columns are:
- Name
- Degree Centrality
- Betweenness Centrality
- Closeness Centrality

Briefly discuss who are the most "central" people in the social network and why.

```python
# Calculate the Degree Centrality of each node/person on the network
degree_centralities = nx.degree_centrality(G)

# Calculate the Betweeness Centrality for each node/person on the network
betweenness_centralities = nx.betweenness_centrality(G)

# Calculate the Closeness Centrality for each node/person on the network
closeness_centralities = nx.closeness_centrality(G)

# Creates a dataframe that summarizes the centrality measurements observed on the network
centrality_table = pd.DataFrame({
    'Node': list(G.nodes()),
    'Degree Centrality': list(degree_centralities.values()),
    'Betweenness Centrality': list(betweenness_centralities.values()),
    'Closeness Centrality': list(closeness_centralities.values()),
})

# Shows the dataframe
centrality_table
```

There are various centrality measures that describes a network in many different ways. However, for this table, we only plan to tackle three specific measurements, namely: Degree, Betweeness and Closeness Centralities.

1. Degree Centrality = "Who knows the most actors?"

  This measurement of a network describes a node's or a person's connection to other nodes/people. It is also defined by the number of links or direct neighbors connected to a node. The higher value of degree centrality a node has, the more connections it has on the network.

  The formula for the degree centrality $C_D$ is given by

  \begin{equation}
      C_D(n_i) = \frac{d(n_i)}{g-1}
  \end{equation}
  
  where $d$ is the number of links connected to the node and $g$ is the number of nodes in the network.

  The person with the highest degree centrality is Jill, with a value of 0.66667. This makes her the most popular person in their friend group.

2. Betweenness Centrality = "Who controls knowledge flow?"

  This measures a node's position in the middle of the middle of a network, specifically its location in the shortest path between nodes. The higher value of betweenness centrality a node has, the more centered its position is in the network.

  The formula for betweenness centrality is given by

  \begin{equation}
    C_B(n_i) = \frac{\sum_{j<k} g_{jk}(n_i)}{g_{jk}}
  \end{equation}

  The person with the highest betweenness centrality is Liz, with a value of 0.38889. This makes sense since she seems to be located at the center of the network.

3. Closeness Centrality = "Who has the shortest distance to the other actors?"

  This indicates how close a person/node is from the rest of the people/nodes in the network. The higher value of closeness centrality a node has, the more connected it is from the rest of the nodes.
  
  The formula for closeness centrality is given by

  \begin{equation}
    C_c(n_i) = \frac{1}{\sum_{j=1}d(n_i,n_j)}
  \end{equation}  

  The people with the highest closeness centrality are shared by Shane and Emma, with a value of 0.64286. We can observe that in the network, they seem to share a similar position, which makes them closest to everyone else when compared to other people.

#### PART 3 - Identify communities *(40 points)*

1. Identify communities in the social network using *(30 points)*:
- Louvain community detection algorithm
- Any other community detection algorithm of your choice

2. Calculate the modularity of the partitions from the two community detection algorithms. *(10 points)*

Briefly describe the algorithm you used and compare the results from the two algorithms.

```python
# Creates the list that shows the communities according to Louvain community detection algorithm
partition1 = nx.community.louvain_communities(G)

# Calculates the modularity of the partitions, based on the Louvain community detection algorithm
modularity1 = nx.community.modularity(G, partition1)

# Prints the outputs
print("The list of communities seen, based on the Louvain community detection algorithm, are: " + str(partition1))
print("The calculated modularity of the communities is " + str(modularity1))
```

```python
# Creates the list that shows the communities according to edge betweenness detection algorithm
partition2 = nx.community.edge_betweenness_partition(G, 3)

# Calculates the modularity of the partitions, based on the edge betweenness detection algorithm
modularity2 = nx.community.modularity(G, partition2)

# Prints the outputs
print("The list of communities seen, based on the edge betweenness detection algorithm, are: " + str(partition2))
print("The calculated modularity of the communities is " + str(modularity2))
```

The two community detection algorithms I used are the Louvain community and edge betweenness detection. These two have very different approaches when it comes to creating the communities inside a network.

The Louvain community detection algorithm ensures that it finds and extracts the best community structures for the network. On the other hand, the approach that the edge betweenness detection algorithm uses is by removing the edge with the highest edge betweenness value. We then decide the number of communities the network will be divided at, which is 3, to match the number of communities created from the Louvain algorithm.

The different communities created by the two algorithms are expectedly different since they use different approaches. Similarly, the calculated modularities for these are different as well. By definition, modularity is the density of connections per community. A higher modularity value represents a higher connection within the community, but shows a lower amounts of connection outside that community.

The calculated modularity for the Louvain algorithm is expectedly higher than the edge betweenness algorithm since it was described earlier than the former uses an efficient way to divide the network to the best possible communities.
