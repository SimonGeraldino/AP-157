```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

#### PART 1 - Generate data set *(20 points)*

1. Generate a toy data set with 100 regularly-spaced points and as described by Eq. 8.75-- a simple model where $x$ and $y$ satisfy the following:
$$ 0 \le x_i \le 3 $$
$$ y_i = x_i \sin(x_i) + \epsilon_i $$
where the noise is drawn from a normal distribution $\epsilon_i \sim \rm{N}(0, 0.1)$. *(10 points)*  

2. Plot your toy data set (Refer to Fig. 8.12 for comparison). *(10 points)*


```python
"""
We are tasked to create a toy data set, with 100 regularly-space points.
A random noise is added in each of these points. The toy data set will be shown.
"""

# We create array x, which is a list of numbers from 0 to 3, with 100 regularly-spaced points.
x = np.linspace(0,3,100)

# We create the random varying variable epsilon as the "noise" that is added in the data points. Its value changes from 0 to 0.1.
epsilon = np.random.normal(0,0.1,100)

# We define this function, according to Eq. 8.75 from the SDMMLA reference.
def f(x):
    return x * np.sin(x) + epsilon

# We put all the values of the function of x in y.
y = f(x)

# This is the visual representation of the toy data set, in the form of a scatter plot.
plt.scatter(x, y)
plt.title('Toy data set')

# This shows the plot.
plt.show()
```

# PART 2 - Apply cross-validation *(80 points)*

Recreate the top panel of Figure 8.14: rms error vs. polynomial degree for the training set and cross-validation set. To do this, you will perform the ff. steps:

1. Split the data set into training, cross-validation, and test sets with 50%, 25% and 25% ratio. You can use sklearn library's model_selection.train_test_split function [(link)](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html). *(20 pts.)*
2. Get the best-fit curve for polynomial degrees $d=0$ to 14 for the training set. You can use numpy library's polyfit function [(link)](https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html). *(30 pts.)*
3. Get the rms errors for both the training and cross-validation sets (for the best-fit curve from Step 2). *(10 pts.)*
4. Plot rms errors for both the training and cross-validation sets against polynomial degree $d$ (as in Fig. 8.14). *(20 pts.)*


```python
"""
We want to split the data set into training, cross-validation and test sets.
50% of the set will be "given" to the training set and 25% will be "given" for both the testing and cross-validation sets.

The function used for this is the train_test_split function from the sklearn library.
A random state, given by the value in the random_state parameter, will be added to the function,
but this should not affect the essence of what the function does.
"""

from sklearn.model_selection import train_test_split

# The data set is divided to 50% training set and 50% "testing" set.
x_train, x_test1, y_train, y_test1 = train_test_split(x, y, train_size = 0.50, test_size = 0.50, random_state = 42)

# The "testing" set from the first use of train_test_split function will be then equally divided for the testing and
# cross-validation sets. This makes them have 25% of the original set, for each set.
x_test2, x_xval, y_test2, y_xval = train_test_split(x_test1, y_test1, train_size = 0.50, test_size = 0.50, random_state = 42)
```

```python
# We create empty lists ploy_list for the polynomial fit list.
poly_list = []

# Since we were asked to create best-fit curves, from polynomial degrees 0 to 14, we will create an array list for it.
deg = np.arange(0,15,1)

# This will be the "for" loop that creates the best-fit curves, then graphs it.
for i in deg:

    # With this, we make a function that constructs a polynomial curve with the least squares.

    # np.polyfit = least squares polynomial fit
    # np.poly1d = constructs the polynomial
    model = np.poly1d(np.polyfit(x_train, y_train, i))

    # This appends the curve that is created with every iteration of the array x.
    poly_list.append(model(x))

    # Graphs the created polynomial curves from the "poly-list" array.
    plt.plot(x,poly_list[i])
    plt.title("Best Fit Line")
    plt.xlabel("X")
    plt.ylabel("Y")
```

```python
from sklearn import metrics

# We will create these empty array lists for the best fit lines.
train_pfit = []
xval_pfit = []

# Creates empty array lists for root mean square errors
train_rmse = []
xval_rmse = []

# This for loop will be used to create the list of best fit lines that will be visually shown.
for i in deg:

  model = np.poly1d(np.polyfit(x_train, y_train, i))
  train_pfit.append(model(x_train))
  xval_pfit.append(model(x_xval))

# This for loop will be used to create the list of root mean square errors that will be visually shown.
for j in deg:

  train_rmse.append(metrics.mean_squared_error(y_train, train_pfit[j], squared = False))
  xval_rmse.append(metrics.mean_squared_error(y_xval, xval_pfit[j], squared = False))
```

```python
# We will now visually represent the root mean square errors for both training and cross-validation lists against the polynomial degree list.

# Plots the root mean square error for the training set.
plt.plot(deg, train_rmse, label = "Training set")

# Plots the root mean square error for the cross-validation set.
plt.plot(deg, xval_rmse, label = "Cross-validation set")
plt.title("Root Mean Square Errors vs Polynomial Degree")
plt.xlabel("Degree")
plt.ylabel("RMS Error")
plt.legend()

# This shows the plot.
plt.show()
```
