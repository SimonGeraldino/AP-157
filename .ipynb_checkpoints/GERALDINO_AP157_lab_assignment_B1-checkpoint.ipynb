#### PART 1 - Create synthetic greyscale images *(30 points)*

Simulate a 1 cm x 1 cm optical element with a sinusoid along the x-direction with a given frequency-- generate images with frequencies of 1, 3, and 5 cycles/cm. The size of each image should be 400x400 pixels.

For each frequency, show the:
1. [15 pts.; 5 pts. each] Image heatmap (range of pixel values should be between 0 to 255)
2. [15 pts.; 5 pts. each] Plot showing 3-D view



Sample output for frequency of 5 cycles/cm is uploaded to the Google Drive Datasets folder: https://drive.google.com/file/d/1770FiH4FPm0C_fCpRGorr64x-RIv9UOY/view?usp=drive_link



```python
# Sets the variables to be used
freq = [1, 3, 5]  #cycles/cm
grid_width = 400  #pixels
image_width = 1   #cm

# Creates the meshgrid
x = np.linspace(0, image_width, grid_width)
y = x
X, Y = np.meshgrid(x, y)

# Sets the size of the figure
fig = plt.figure(figsize=(10,5))

# Creates a for loop that generates the images based on the function given
for i in range(0,3):

  # Creates the sinusoidal function
  sinusoid = np.sin(2 * np.pi * freq[i] * X) * 255

  # Sets up the subplots for the 2D image
  plt.subplot(2, 3, i+1)

  # Shows the image created from the function
  plt.imshow(sinusoid, cmap="magma")

  # Adds parameters on the 2D images
  plt.title("Frequency = " + str(freq[i]) + " Hz")
  plt.axis("off")

  # Adds the subplots for the 3D image
  ax = fig.add_subplot(2, 3, i+4, projection='3d')

  # Plots the 3D sinusoidal function
  ax.plot_surface(X, Y, sinusoid, cmap="magma")

  # Adds parameters on the 3D images
  ax.set_title("3D Sinusoidal Plot of Frequency = " + str(freq[i]) + " Hz", fontsize = 10.5)
  ax.grid(True)

# Shows the plot
plt.tight_layout()
plt.show()
```

We were asked to simulate 2D and 3D images to represent a sinusoidal function, that has varying frequency. The 1 by 1 cm image was divided into 400 by 400 pixels.

The resulting images show what was expected from a sinusoidal function with a specified frequency. For example, for 5 Hz frequency, the 2D image shows 5 light and dark bands each, which represent the crests and troughs of each of the wave. (Frequency represents the number of waves/cycles per unit of measurement. In this case, 5 Hz = 5 waves/cm and the width of the optical element = 1 cm, which meant that we will only see 5 waves in the image).

#### PART 2 - Create color images *(30 points)*

Recreate the Olympics logo as an image. It is fine if the colors overlap where the rings intersect.

Sample output is uploaded to the Google Drive Datasets folder: https://drive.google.com/file/d/1770FiH4FPm0C_fCpRGorr64x-RIv9UOY/view?usp=drive_link

```python
# Sets the variables to be used
grid_width = 1024  #pixels
inner_radius = 2.4
outer_radius = 2.8
deg = 45

# Creates the meshgrid
x = np.linspace(-10, 10, grid_width)
y = x
X, Y = np.meshgrid(x, y)

# Initializes the color arrays
Red = np.zeros((grid_width, grid_width))
Blue = np.zeros((grid_width, grid_width))
Green = np.zeros((grid_width, grid_width))

# Calculates the coordinates of the center of the colored circles
x_coords = outer_radius * np.cos(deg * np.pi / 180)
y_coords = outer_radius * np.sin(deg * np.pi / 180)

# Creates the colored annuli

# Green Annulus
R = np.sqrt((X - 1.6*x_coords) ** 2 + (Y - 1.2*y_coords) ** 2)
Red[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0
Green[np.where((R < outer_radius) & (R >= inner_radius))] = 0.0
Blue[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0

# Yellow Annulus
R = np.sqrt((X + 1.6*x_coords) ** 2 + (Y - 1.2*y_coords) ** 2)
Red[np.where((R < outer_radius) & (R >= inner_radius))] = 0.0
Green[np.where((R < outer_radius) & (R >= inner_radius))] = 0.0
Blue[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0

# Blue Annulus
R = np.sqrt((X + 3.2*x_coords) ** 2 + (Y) ** 2)
Red[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0
Green[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0
Blue[np.where((R < outer_radius) & (R >= inner_radius))] = 0.0

# Black Annulus
R = np.sqrt((X) ** 2 + (Y) ** 2)
Red[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0
Green[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0
Blue[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0

# Red Annulus
R = np.sqrt((X - 3.2*x_coords) ** 2 + (Y) ** 2)
Red[np.where((R < outer_radius) & (R >= inner_radius))] = 0.0
Green[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0
Blue[np.where((R < outer_radius) & (R >= inner_radius))] = 1.0

# Creates a 3D array to represent the RGB image
I_rgb = np.zeros((grid_width, grid_width, 3))
I_rgb[..., 0] = Red
I_rgb[..., 1] = Green
I_rgb[..., 2] = Blue

# Adds parameters to the image and grid
plt.figure(figsize=(5, 5), dpi=150)
plt.grid(color='grey', linestyle='--', linewidth=0.5)  # Add grid
plt.title("Olympics Rings", fontsize=15)

# Plots the RGB image
plt.imshow(1-I_rgb)

# Shows the image
plt.show()
```

For this part, we were asked to recreate the "Olympics Rings" logo = colored rings + white background. We needed annuli for the rings and different colors to copy the color for each of the rings. The first step required creating multiple circles, with the same radius, and subtracting smaller circles, with a smaller radius. Then, the three main colors (Red, Green & Blue) were given values, according to the color needed in each of the rings. However, to recreate the logo, we needed to get the inverse since the image made in the 'I_rgb' array will produce a black background. This meant that the color values that we needed to set per ring must also be opposite of the actual colors for the logo. Last step required us to rearrange the rings to their relative positions. After these steps, we now have recreated the "Olympics Rings" logo.

#### PART 3 - Apply image enhancement *(40 points)*

Apply contrast stretching to an astronaut image.

Show the following:
1. [20 pts.] Original and contrast-stretched image
2. [10 pts.] Probability Distribution Function (PDF) plot showing curves for the original and contrast-stretched image
3. [10 pts.] Cumulative Distribution Function (CDF) plot showing curves for the original and contrast-stretched image

Sample output is uploaded to the Google Drive Datasets folder: https://drive.google.com/file/d/12-cPwAr7hVUPBKIthvcQgvqjvEOmwyxJ/view?usp=drive_link

First step is to load the image-- available from the scikit-image (skimage) library-- and make a low-contrast version, using the code below:

```python
from skimage import color, data, exposure
from skimage.transform import rescale

img = color.rgb2gray(data.astronaut())*256 # original astronaut image
plt.figure(figsize=(10,5), dpi = 200)
plt.subplot(121)
plt.imshow(img, cmap = 'gray', vmax = 255)
plt.colorbar()

img_low_contrast = img/5 # low contrast astronaut image
plt.subplot(122)
plt.imshow(img_low_contrast, cmap = 'gray', vmax = 255)
plt.colorbar()
```

Next, write your code to apply contrast stretching, calculate the PDF and CDF of the original and contrast-stretched images. Finally, display the images and PDF and CDF plots.

```python
# Creates a function for histogram analysis for the original image

# Sets the variables to be used
N = 256 #pixels

# Calculates the histogram
hist1, bins1 = np.histogram(img.flatten(), bins = N, range=[0,N])

# Creates the cumulative distributive function
cdf1 = hist1.cumsum()

# Normalizes the CDF
cdf1_normalized = cdf1 / cdf1.max()

# Transforms image to 1D values
image_1D = img.flatten()

# Creates a function for histogram analysis for the contrast stretched image (using the same method)
hist2, bins2 = np.histogram(img_low_contrast.flatten(), bins = N, range=[0,N])
cdf2 = hist2.cumsum()
cdf2_normalized = cdf2 / cdf2.max()

# Displays both the original and low contrast images
fig,ax = plt.subplots(1, 2, figsize=(8,4), dpi=150)
ax[0].imshow(img_low_contrast, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[1].imshow(img, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Histogram Manipulated')

# Plots the Probability Distribution Function for both the images
fig, ax = plt.subplots(1, 2, figsize=(8, 4), dpi=150)

# Transforms image to 1D values
ax[0].hist(img_low_contrast.flatten(), 256, [0, 256], color='r')
ax[0].hist(img.flatten(), 256, [0, 256], color='b')

# Sets the parameters for the PDF graph
ax[0].set_title('Probability Distribution Function')
ax[0].set_xlim([0,255])
ax[0].legend(('Original', 'Contrast Stretched'), fontsize=8)

# Plots the Cumulative Distribution Function for both the images
ax[1].plot(cdf2_normalized, 'b')
ax[1].plot(cdf1_normalized, 'r')

# Sets the parameters for the CDF graph
ax[1].set_title('Cumulative Distribution Function')
ax[1].set_xlim([0,255])
ax[1].legend(('Original', 'Contrast Stretched'), fontsize=8)

# Shows the images and the graphs
plt.tight_layout()
plt.show()
```
